```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from matplotlib import rc
rc("text", usetex=False)
```

```python
data = pd.read_csv("2017 SSBM Head to Head - Top 50 H2H.csv")
```

```python
data.head(5)
```

```python
rank2016 = np.array(data["2016"])
rank2017 = np.array(data["2017"])
player = np.array(data["Unnamed: 2"])
        
def isNaN(num):
    return num != num
```

```python
print(player)
```

```python
my_dict = {}
for thisPlayer in player[0:15]:
    my_dict[thisPlayer] = []
    for index, row in data.iterrows():
            
        if "-" in str(row[thisPlayer]):
                        
            setCounts = row[thisPlayer].replace(" ","")
            setCounts = row[thisPlayer].replace("*","")
            
            losses = np.float(setCounts.split("-")[0])
            wins = np.float(setCounts.split("-")[1])
            
            if losses == 0.0:
                datapoint = [100.0, player[index]]
                my_dict[thisPlayer].append(datapoint)
            else:
                if wins == 0.0:
                    datapoint = [-100.0, player[index]]
                    my_dict[thisPlayer].append(datapoint)
                if wins > losses:
                    datapoint = [wins / losses * 100.0, player[index]]
                    my_dict[thisPlayer].append(datapoint)
                if wins < losses:
                    datapoint = [wins / losses * -100.0, player[index]]
                    my_dict[thisPlayer].append(datapoint)
                            
        if isNaN(row[thisPlayer]) == True:
            datapoint = [" ", player[index]]
            my_dict[thisPlayer].append(datapoint)
```

```python
for key,value in my_dict.items():
    if key == "Leffen":
         print(value)
```

```python
for key,value in my_dict.items():
    if key == player[0]:
        plt.plot([1,2,3],[1,2,3])
        plt.figure(figsize=(11,7.5))
    for i in value:
        if type(i[0]) != str:
            plt.bar(i[1], i[0], color = 'green' if i[0] > 0 else 'red')
        else:
            plt.bar(i[1], i[0])
    plt.title(key)
    plt.xticks(rotation=90)
    plt.yticks([-200,-150,-100,-50,0,50,100,150,200], labels=[" ",-300,-200,-100,0,100,200,300])
    plt.xlabel('Opponent')
    plt.ylabel('Win %')
    if key in player[0:10]:
        saveString = '2017MeleePlots/' + str(key)
        plt.savefig(saveString)
    plt.figure(figsize=(11,7.5))
```
